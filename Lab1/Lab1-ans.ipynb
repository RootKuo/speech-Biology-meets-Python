```python
%matplotlib inline
import matplotlib.pyplot as plt
file = open("data/dna.txt", "r")
print("Input data is :")
for columns in ( raw.split() for raw in file ):  
    print(columns[0])
DNA = columns[0]

fig1= plt.figure(figsize=(6, 6))
ax = fig1.add_axes([0.1, 0.1, 0.8, 0.8])

def count_ATCG(input_string):
    # You need to add a 1d array to count A, T, C, G  
    ans = [input_string.count("A"),
           input_string.count("T"),
           input_string.count("C"),
           input_string.count("G")]
    return ans

key = 'Adenine', 'Thymine', 'Cytosine', 'Guanine'
value = count_ATCG(DNA)

jump = (0, 0.1, 0, 0)
color = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

pies = ax.pie(value, explode=jump, labels=key, autopct='%1.1f%%', colors=color )
plt.show()
```
